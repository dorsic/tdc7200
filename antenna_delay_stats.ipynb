```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import allantools
```

```python
df = pd.read_csv('/Users/dorsic/Documents/PhD/TDC7200/antenna_delay9.csv')
```

```python
N_points = len(df)
n_bins = 50

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(df['tof']*1e9, bins=n_bins)
plt.xlabel('ns')
axs[1].hist(df['tof']*1e9, bins=2*n_bins)
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.title('TOF (time of flight) = Time Interval [nanoseconds]')
plt.xlabel('sec')
plt.ylabel('ns')
plt.plot(df['tof']*1e9)
```

```python
a = allantools.Dataset(data=df['tof'])
a.compute("oadev")
b = allantools.Plot()
b.plot(a, errorbars=True, grid=True)
b.ax.set_xlabel("Tau (s)")
b.show()
```
